## Splitting the data 

```python
X= heart_df.values[:,:-1]
Y= heart_df.values[:,-1]
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2,
                                                   random_state=10)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
Y_train= Y_train.astype(int)
Y_test= Y_test.astype(int)
```

## Scaling the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X_train)
X_train= scaler.transform(X_train)
X_test= scaler.transform(X_test)
```

```python
print(X_train)
print(X_test)
```

## Model Building

```python
from sklearn.linear_model import LogisticRegression

blr_model= LogisticRegression()

blr_model.fit(X_train, Y_train)
Y_pred= blr_model.predict(X_test)
print(list(zip(Y_test,Y_pred)))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

print("Confussion Matrix: \n", confusion_matrix(Y_test, Y_pred))
print("Accuracy Score :\n", accuracy_score(Y_test,Y_pred))
print("classification_reprt :\n", classification_report(Y_test,Y_pred))
```

```python
Y_probability = blr_model.predict_proba(X_test)
print(Y_probability)
```

```python
for i in np.arange(0.4,0.71,0.01):
    predict_mine = np.where(Y_probability[:,1]>i,1,0)
    cfm=confusion_matrix(Y_test,predict_mine)
    total_error = cfm[0,1]+cfm[1,0]
    print("Error at threshold: ", i," : ", total_error, "Type_1 Error : ",cfm[0,1],
          "Type_2 Error : ", cfm[1,0] )
```

```python
Y_pred_class=[]
for i in Y_probability[:,1]:
    if i>0.65:
        Y_pred_class.append(1)
    else:
        Y_pred_class.append(0)
print(Y_pred_class)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

print("Confussion Matrix: \n", confusion_matrix(Y_test, Y_pred_class))
print("Accuracy Score :\n", accuracy_score(Y_test,Y_pred_class))
print("classification_reprt :\n", classification_report(Y_test,Y_pred_class))
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=15, metric="euclidean")
knn_model.fit(X_train, Y_train)
Y_pred= knn_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

print("Confussion Matrix: \n", confusion_matrix(Y_test, Y_pred))
print("Accuracy Score :\n", accuracy_score(Y_test,Y_pred))
print("classification_reprt :\n", classification_report(Y_test,Y_pred))
```

```python
my_dict = {}
for i in range (1,16):
    knn_model = KNeighborsClassifier(n_neighbors=i, metric="euclidean")
    knn_model.fit(X_train, Y_train)
    Y_pred= knn_model.predict(X_test)
    print("Accuracy : ", accuracy_score(Y_test,Y_pred), "for K value: ", i)
    my_dict[i]=accuracy_score(Y_test,Y_pred)
    
```
